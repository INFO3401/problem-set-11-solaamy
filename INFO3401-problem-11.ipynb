```python
import pandas as pd
import numpy as np
import altair as alt

# Must use for Jupyter Notebook
alt.renderers.enable('notebook')
```

```python
import csv
df = pd.read_csv("sentimentData.csv")
df
```

```python
print(df.columns)
```

```python
#code for prob.5
alt.Chart(df).mark_point().encode(
    x='Author',
    y='Positive'
).interactive()

##From the graph, Shakespare has relatively more positive sentiment than Marlowe. 
##While Marlowe has a point with highest value of positive sentiment, Shakespare has a point with lowest value of positive sentiment.
##The distribution of positive sentiment in Shakespare is evenly distributed. The distribution of positive sentiment in Marlowe is mainly between 0.10-0.15.
```

```python
df_raw = pd.read_csv("rawCounts.csv")
df_raw
```

```python
print(df_raw.columns)
```

```python
groupData = df_raw.groupby(['Author','Word']).sum()
sorts = groupData.reset_index().sort_values(['Author','Count'],ascending=False).set_index(['Author','Word'])
print(sorts)
head_2500 = sorts.groupby('Author').head(2500).reset_index()
print(head_2500)
```

```python
#code for prob.6
alt.Chart(head_2500).mark_point().encode(
    x='Author',
    y='Count',
).interactive()
#The top 250 words counts in Marlowe's plays mainly distribute between 0-4,000 and only one word count more than 4,000. 
#The top 250 words counts in Shakespeare's plays mainly distribute between 0-4,000 but it also have severl words that counts more than 4,000 
```

```python
#code for prob.7
#
#Visualization.1
alt.Chart(df).mark_bar().encode(
    x='Author',
    y='average(Negative)',
).interactive()
#Compared to the average value of negative sentiment, we can see Shakespare's plays appear more negative sentiment than Marlowe's plays.
```

```python
#Visualization.2
alt.Chart(df).mark_point().encode(
    x='Positive',
    y='Negative',
    color = 'Author'
).interactive()
#Shakespeare's plays relatively appears more positive and negative sentiment than Marlowe's plays. 
#Also, one of the Marlowe's plays showest the highest positive sentiment. The value is around 0.19.
```

```python
#Visualization.3
head_250 = sorts.groupby('Author').head(250).reset_index()
alt.Chart(head_250).transform_window(
    rank='rank(Count)',
    sort=[alt.SortField('Count', order='descending')]
).transform_filter(
    alt.datum.rank < 100
).mark_bar().encode(
    y=alt.Y('Word:N',
        sort=alt.EncodingSortField(field='Count', op='sum', order='descending')
    ),
    x='Count:Q',
    color = 'Author'
)
#we can see the word that appears the most in both authors' plays is "the". Compared the words that have most counts, these words appear more in Shakespeare's plays than Marlowe's plays.
```
